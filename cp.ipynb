```python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset
file_path = "data_smartphones.xlsx"  # Pastikan file berada di direktori yang benar
data = pd.read_excel(file_path)

# Mengonversi kolom Ratings ke tipe numerik jika masih berbentuk string
data['Ratings'] = pd.to_numeric(data['Ratings'], errors='coerce')

# Memfilter smartphone dengan rating tertinggi dan mengambil minimal 10 data
top_rated_smartphones = data.nlargest(10, 'Ratings')[['Type', 'Brand', 'Price', 'Ratings']]

# Membuat diagram batang horizontal
plt.figure(figsize=(14, 8))
bars = plt.barh(top_rated_smartphones['Type'], top_rated_smartphones['Ratings'], color='lightblue')

# Menambahkan judul dan label
plt.title('10 Smartphone dengan Rating Tertinggi', fontsize=16, weight='bold')
plt.xlabel('Rating', fontsize=14)
plt.ylabel('Tipe Smartphone', fontsize=14)
plt.gca().invert_yaxis()  # Membalikkan urutan agar rating tertinggi di atas
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Menambahkan nilai rating di samping batang
for bar, rating in zip(bars, top_rated_smartphones['Ratings']):
    plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2, f'{rating:.1f}', 
             va='center', ha='left', fontsize=12, color='black')

# Menampilkan diagram
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Memuat dataset
file_path = "data_smartphones.xlsx" 
data = pd.read_excel(file_path)

# Memilih dan membersihkan data
data = data[['Brand', 'Price']].dropna()
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Mengelompokkan berdasarkan merek dan mengambil harga tertinggi dari setiap merek
top_brands = data.groupby('Brand')['Price'].max().nlargest(10).reset_index()

# Membuat Diagram Batang
plt.figure(figsize=(14, 8))
plt.bar(top_brands['Brand'], top_brands['Price'], color='royalblue')

# Menambahkan judul dan label
plt.title('10 Merek HP dengan Harga Tertinggi', fontsize=16, weight='bold')
plt.xlabel('Merek HP', fontsize=14)
plt.ylabel('Harga Tertinggi (Rp)', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan plot
plt.show()
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler

# Memuat dataset
file_path = "data_smartphones.xlsx"  # Pastikan lokasi file sesuai
data = pd.read_excel(file_path)

# Memilih fitur yang digunakan untuk perhitungan similarity
features = ['Price', 'Ratings']

# Mengonversi ke tipe numerik dan menghapus data kosong
data[features] = data[features].apply(pd.to_numeric, errors='coerce')
data = data.dropna(subset=features).reset_index(drop=True)


data = data.iloc[:711]

# Normalisasi data agar berada dalam rentang [0, 1]
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(data[features])

# Menghitung cosine similarity
similarity_matrix = cosine_similarity(normalized_data)

# Membuat DataFrame hasil similarity
similarity_df = pd.DataFrame(similarity_matrix, index=data['Type'], columns=data['Type'])

# Menampilkan hanya 200 smartphone pertama
print("Matrix Similarity Smartphone :")
print(similarity_df)
```

```python
#MODELING
# Input pengguna
input_name = input("Masukkan nama hp yang diinginkan (contoh: realme): ").lower()  # Mengubah input ke huruf kecil
input_rom = int(input("Masukkan kapasitas RAM yang diinginkan (dalam GB): "))
input_ram = int(input("Masukkan kapasitas ROM yang diinginkan (dalam GB): "))

# Content-Based Filtering
# TF-IDF Vectorization untuk nama
tfidf = TfidfVectorizer()
name_vectors = tfidf.fit_transform(data['Type'].str.lower())

# Gabungkan fitur TF-IDF dengan ROM, RAM, dan Ratings
numerical_features = data[['RAM (GB)', 'ROM (GB)', 'Ratings']].values
combined_features = pd.concat([
    pd.DataFrame(name_vectors.toarray()),
    pd.DataFrame(numerical_features)
], axis=1).values

# Query pengguna
query_vector = tfidf.transform([input_name]).toarray()
query_numerical = [[input_rom, input_ram, 0]]  # Ratings diatur ke 0 untuk pencarian
query_combined = pd.concat([
    pd.DataFrame(query_vector),
    pd.DataFrame(query_numerical)
], axis=1).values

# Hitung kesamaan menggunakan cosine similarity
similarities = cosine_similarity(query_combined, combined_features)

# Tambahkan skor kesamaan ke dataset
data['Similarity'] = similarities.flatten()

# Filter hasil berdasarkan nama, ROM, dan RAM
data_filtered = data[
    (data['Type'].str.lower().str.contains(input_name)) &  # Filter nama berbasis input
    (data['RAM (GB)'] == input_rom) &
    (data['ROM (GB)'] == input_ram)
]

# Periksa apakah ada data yang cocok
if not data_filtered.empty:
    # Urutkan hasil berdasarkan kesamaan tertinggi
    recommended = data_filtered.sort_values(by='Similarity', ascending=False)
    
    # Tampilkan hasil dalam tabel rapi menggunakan tabulate
    print("\nHasil Rekomendasi:")
    print(tabulate(recommended[['Brand', 'Type', 'RAM (GB)', 'ROM (GB)', 'Ratings', 'Price', 'Camera', 'Battery', 'Similarity']],
                   headers='keys', tablefmt='fancy_grid'))
else:
    print("\nTidak ada smartphone yang memenuhi kriteria pencarian Anda.")
    
    
# EVALUASI
# Ground Truth: Mark item relevan (1) jika sesuai dengan input pengguna
y_true = [
    1 if (row['Type'].lower().startswith(input_name) and 
          row['RAM (GB)'] == input_rom and 
          row['ROM (GB)'] == input_ram)
    else 0
    for _, row in data.iterrows()
]

# Prediksi: Mark item relevan (1) jika masuk dalam hasil filter
y_pred = [1 if idx in data_filtered.index else 0 for idx in data.index]

# Hitung Precision, Recall, dan F1-Score
precision = precision_score(y_true, y_pred, zero_division=0)
recall = recall_score(y_true, y_pred, zero_division=0)
f1 = f1_score(y_true, y_pred, zero_division=0)

# Cetak hasil evaluasi
print("\nEvaluasi Sistem Rekomendasi:")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```
